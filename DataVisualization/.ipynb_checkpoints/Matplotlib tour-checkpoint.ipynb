```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
np.random.seed(19680801)
```

%matplotlib inline

```python
data = np.arange(10)
plt.plot(data)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
ax3.plot(np.random.standard_normal(50).cumsum())
```

```python
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)
axes[0, 0].scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
fig.subplots_adjust(wspace=0.1, hspace=0.1)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)
```

O percentual de largura e altura entre as subplotagens podem ser manipuladas através dos parâmetros **wspace** e **hspace**, respectivamente, do método subplots_adjust() da classe Figure.

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(30).cumsum(), color="black", linestyle="dashed", marker='o')
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data, color="black", linestyle="dashed", label="Default")
ax.plot(data, color="black", drawstyle="steps-post", label="Steps-Post")
ax.legend()
```

O parâmetro drawstyle do método plot determina a forma como os subsquentes são interpolados. Por padrão, eles são interpolados linearmente. 

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=30, fontsize=8)
ax.set_xlabel("Stages")
ax.set_ylabel("Steps")
ax.set_title("Not my first and not my last")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(1000).cumsum(), color="red", label="one")
ax.plot(np.random.standard_normal(1000).cumsum(), color="green", linestyle="dashed", label="two")
ax.plot(np.random.standard_normal(1000).cumsum(), color="blue", linestyle="dashed", label="three")
ax.legend()
```

```python
# Fixing random state for reproducibility
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
colors = ['red', 'green', 'blue', 'orange', 'yellow']
ax.barh(y_pos, performance, xerr=error, align='center', color=colors)
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
with plt.xkcd():
    np.random.seed(19680801)
    data = np.ones(100)
    data[60:] += np.arange(40)
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_ylabel("Grades")
    ax.set_xlabel("Exams")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.plot(data)
    ax.annotate("After breaking up with my girlfriend", xy=(60, 1), arrowprops=dict(arrowstyle='->'), xytext=(2, 30))
```

with.xkcd() plota um gráfico estilo xkcd. A fonte pode ser baixada aqui https://github.com/ipython/xkcd-font/blob/master/xkcd-script/font/xkcd-script.ttf.  ax.spines[['top', 'right']].set_visible(False) deixa invisíveis as "espinhas" do lado superior, 'top', e direito, 'right'. ax.annotate cria uma anotação dentro da figura.
